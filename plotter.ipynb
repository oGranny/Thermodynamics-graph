```python
import numpy as np
import matplotlib.pyplot as plt
```

## Here all the constants are defined

<p style="text-align: center; font-size: 24px;">aA + bB &rarr; cC</p>

#### here are what all the constants are:
- `COEF_A` is value of a
- `COEF_B` is value of b
- `COEF_C` is value of c
- `DELTA_G` is value of △G°
- `DELTA_H` is value of △H°
- `TEMP_START` is the starting value of temperature range
- `TEMP_END` is the starting value of temperature range
- `TEMP_STEP` is the increment that will be done till TEMP_END is reached


```python
COEF_A = 1
COEF_B = 1
COEF_C = 1
DELTA_G = -5000
DELTA_H = 5000
TEMP_START = 300
TEMP_END = 1000
TEMP_STEP = 100
# PRESSURE = 1
R = 8.314
```

```python
Na = np.linspace(0,1,600)
```

##### All the mole fraction are declared below

```python
Xa = (1-COEF_A  * Na) / (2+ (COEF_C - COEF_B - COEF_A) * Na)
Xb = (1-COEF_B  * Na) / (2+ (COEF_C - COEF_B - COEF_A) * Na)
Xc = (COEF_C  * Na) / (2+ (COEF_C - COEF_B - COEF_A) * Na)
```

##### `get_del_g` function calculates the △G° for different Temperatures

```python
def get_del_g(TEMP, Xa, Xb, Xc):
    return R * TEMP * (Xa * np.log(Xa) + Xb * np.log(Xb) + Xc * np.log(Xc))
```

##### △H° is taken as a straight line 

```python
del_H = -1*DELTA_H * Na
```

##### Here loop is excecuted for finding △G° and △H° curves at different temperatures

```python
for TEMP in range(TEMP_START, (TEMP_END+TEMP_STEP), TEMP_STEP):
    del_g = get_del_g(TEMP, Xa, Xb, Xc)
    plt.figure()
    plt.plot(Na, del_H, label='△H°')
    plt.plot(Na, del_g, label='△G°')
    plt.xlabel('Na')
    plt.ylabel('△G°/△H°')
    plt.minorticks_on()
    plt.grid(which='both', linestyle='--', linewidth=0.3)
    plt.grid(True)
    plt.legend()
    plt.title(f'Temperature: {TEMP}K')
    plt.show()
```
